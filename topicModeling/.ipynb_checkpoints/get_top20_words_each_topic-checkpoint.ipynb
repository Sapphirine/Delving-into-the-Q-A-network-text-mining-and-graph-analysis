```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/Zhen/Desktop/Courses/BigData/stackexchange/topicModeling/result/lda_topicMatrix.csv',header = None)
df2 = pd.read_csv('/Users/Zhen/Desktop/Courses/BigData/stackexchange/topicModeling/result/matrix.csv')
```

```python
words = list(df2.columns.values)
```

```python
import json
ntopic=df.shape[1]
nwords=20
topic_frame=pd.DataFrame()
for i in range(ntopic):
    dftmp=df.sort([i], axis=0, ascending=False)
    index1=dftmp.index.tolist()[1:nwords]
    # topic_list.append([(words[j],dftmp.iloc[j,i]) for j in index1])
    topicList=[(words[j],df.iloc[j,i]) for j in index1]
    topicFrame=pd.DataFrame(topicList,columns=['text','size'])
    topic_frame=pd.concat([topic_frame,topicFrame],axis=1)
    # to json
    topicJson=topicFrame.to_json(path_or_buf = None, orient = 'records', date_format = 'epoch', double_precision = 10, force_ascii = True, date_unit = 'ms', default_handler = None)
    # write the file to json
    with open('/Users/Zhen/documents/apache-tomcat-9.0.0.M1/webapps/d3/wordCloud/data/'+str(i)+'.json', "w") as f:
        json.dump(json.loads(topicJson),f,ensure_ascii=False,indent=2)
    f.close()
```

```python
topic_frame.to_csv('/Users/Zhen/Desktop/Courses/BigData/stackexchange/topicModeling/result/topic_word_distribution.csv',sep=',',encoding = 'utf-8',index=False)
```
